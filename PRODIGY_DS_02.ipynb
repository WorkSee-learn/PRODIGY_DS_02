```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the Titanic dataset
titanic_df = pd.read_csv('C:/JUPYTER/Data Files/titanic.csv')

# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Drop 'Cabin' column as it has many missing values and might not be relevant for analysis
titanic_df.drop('Cabin', axis=1, inplace=True)
```

```python
titanic_df.head()
```

```python
# Calculate the median age
median_age = titanic_df['Age'].median()

# Fill missing values with the median age using .loc
titanic_df.loc[titanic_df['Age'].isnull(), 'Age'] = median_age
```

```python
print(median_age)
```

```python
titanic_df
```

```python
titanic_df.drop(['Ticket', 'Fare'], axis=1, inplace=True)
```

```python
titanic_df
```

```python
# Survival count
survival_count = titanic_df['Survived'].value_counts()

# Plot survival count
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_count.index, y=survival_count.values)
plt.title('Survival Count (0 = Not Survived, 1 = Survived)')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()
```

```python
# Survival by age and gender
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Sex', data=titanic_df, multiple='stack', bins=20, kde=True)
plt.title('Survival by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Count')

# Explicitly set labels for legend
plt.legend(title='Sex', labels=['Male', 'Female'])

plt.show()
```

```python
# Survival by passenger class
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
